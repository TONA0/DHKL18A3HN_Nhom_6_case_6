```python
import numpy as np
```

Nhiệm vụ 1: . Đọc dữ liệu khí tượng từ hai file station_bac.csv và station_nam.csv. 

```python
print("\n" + "="*60)
print("NHIỆM VỤ 1: ĐỌC DỮ LIỆU TỪ FILE CSV")
print("="*60)
data_bac = np.genfromtxt('station_bac.csv', delimiter=',', skip_header=1, dtype=str)
print(f"Đã đọc station_bac.csv: {data_bac.shape[0]} dòng")
data_nam = np.genfromtxt('station_nam.csv', delimiter=',', skip_header=1, dtype=str)
print(f"Đã đọc station_nam.csv: {data_nam.shape[0]} dòng")
```

Nhiệm vụ 2: 2. Tách cột date, temp, humidity, rainfall thành các mảng riêng. 

```python
dates_bac = data_bac[:, 0]
temp_bac = data_bac[:, 1]
humidity_bac = data_bac[:, 2]
rainfall_bac = data_bac[:, 3]

dates_nam = data_nam[:, 0]
temp_nam = data_nam[:, 1]
humidity_nam = data_nam[:, 2]
rainfall_nam = data_nam[:, 3]
```

Nhiêm vụ 3:  Chuyển date sang datetime64.

```python
dates_bac_dt = np.array(dates_bac, dtype='datetime64') 
dates_nam_dt = np.array(dates_nam, dtype='datetime64') 
```

Nhiệm vụ 4: Kiểm tra dtype, shape, in 5 giá trị đầu/cuối.

```python

stations = [
    ("BẮC", dates_bac_dt, temp_bac, humidity_bac, rainfall_bac),
    ("NAM", dates_nam_dt, temp_nam, humidity_nam, rainfall_nam)
]
for station_name, dates, temp, humidity, rainfall in stations:
    print(f"TRẠM {station_name}")
    print(f"{'='*50}")
    
    for name, data in [("dates", dates), ("temp", temp), ("humidity", humidity), ("rainfall", rainfall)]:
        print(f"\nTên biến: {name}: Kiểu dữ liệu: {data.dtype}, Shape: {data.shape}")
        print(f"5 giá trị đầu: {data[:5]}")
        print(f"5 giá trị cuối: {data[-5:]}")
```

Nhiệm vụ 5: Slicing 30 ngày đầu, giữa, cuối.

```python

stations = [("BẮC", dates_bac_dt), ("NAM", dates_nam_dt)]
for station_name, dates in stations:
    total_days = len(dates)
    mid = total_days // 2 
    print(f"\nTRẠM {station_name}:")
    print("30 ngày đầu:", dates[:30])
    print("30 ngày giữa:", dates[mid-15:mid+15]) 
    print("30 ngày cuối:", dates[-30:])
```

Nhiệm vụ 6: Boolean mask phát hiện bất thường: nhiệt độ >35°C, độ ẩm <30%, mưa >50mm.

```python
print("=== PHÁT HIỆN BẤT THƯỜNG (BOOLEAN MASK) ===")

stations = [("BẮC", temp_bac, humidity_bac, rainfall_bac), 
            ("NAM", temp_nam, humidity_nam, rainfall_nam)]

for station_name, temp, humidity, rainfall in stations:
    temp_float = np.array(temp, dtype=float)
    humidity_float = np.array(humidity, dtype=float)
    rainfall_float = np.array(rainfall, dtype=float)
    
    print(f"\n{station_name} - Số ngày bất thường:")
    print(f"  Nhiệt độ >35°C: {np.sum(temp_float > 35)}")
    print(f"  Độ ẩm <30%: {np.sum(humidity_float < 30)}")
    print(f"  Mưa >50mm: {np.sum(rainfall_float > 50)}")
```

Nhiệm vụ 7: Broadcasting so sánh nhiệt độ/độ ẩm giữa 2 trạm. 

```python
temp_diff = np.array(temp_nam, dtype=float) - np.array(temp_bac, dtype=float)
humidity_diff = np.array(humidity_nam, dtype=float) - np.array(humidity_bac, dtype=float)

print("SO SÁNH NHIỆT ĐỘ (Nam - Bắc):")
print(f"Chênh lệch trung bình: {np.mean(temp_diff):.2f}°C")
print(f"Số ngày Nam nóng hơn Bắc: {np.sum(temp_diff > 0)}")
print("\nSO SÁNH ĐỘ ẨM (Nam - Bắc):")
print(f"Chênh lệch trung bình: {np.mean(humidity_diff):.2f}%")
print(f"Số ngày Nam ẩm hơn Bắc: {np.sum(humidity_diff > 0)}")
```

Nhiệm vụ 8: Fancy indexing lấy top-10 ngày nhiệt độ cao và mưa lớn nhất mỗi trạm.

```python
stations = [("BẮC", dates_bac_dt, temp_bac, rainfall_bac),
            ("NAM", dates_nam_dt, temp_nam, rainfall_nam)]

for name, dates, temp, rain in stations:
    temp_f, rain_f = np.array(temp, dtype=float), np.array(rain, dtype=float)

    temp_h = np.argsort(temp_f)[-10:][::-1]
    rain_h = np.argsort(rain_f)[-10:][::-1]
    
    print(f"\n{name}:")
    print("10 ngày có nhiệt độ cao nhất:", *[f"{dates[i]}({temp_f[i]:.1f}°C)" for i in temp_h], sep="\n  ")
    print("10 ngày có lượng mưa lớn nhất:", *[f"{dates[i]}({rain_f[i]:.1f}mm)" for i in rain_h], sep="\n  ")
```

Nhiệm vụ 9:  np.diff để phân tích biến động mạnh (ngưỡng: temp >8°C, humidity >15%, rain >40mm).

```python
stations = [("BẮC", temp_bac, humidity_bac, rainfall_bac),
            ("NAM", temp_nam, humidity_nam, rainfall_nam)]

for name, temp, humidity, rain in stations:
    temp_f, humidity_f, rain_f = np.array(temp, dtype=float), np.array(humidity, dtype=float), np.array(rain, dtype=float)
    

    temp_diff = np.abs(np.diff(temp_f))
    humidity_diff = np.abs(np.diff(humidity_f))
    rain_diff = np.abs(np.diff(rain_f))
    
   
    temp_count = np.sum(temp_diff > 8)
    humidity_count = np.sum(humidity_diff > 15)
    rain_count = np.sum(rain_diff > 40)
    
    print(f"\n{name}:")
    print(f"  Nhiệt độ biến động >8°C: {temp_count} lần")
    print(f"  Độ ẩm biến động >15%: {humidity_count} lần")
    print(f"  Mưa biến động >40mm: {rain_count} lần")
```

Nhiệm vụ 10:  Thống kê tổng hợp mean, std, max, min cho từng trạm.

```python

stations = [("BẮC", temp_bac, humidity_bac, rainfall_bac),
            ("NAM", temp_nam, humidity_nam, rainfall_nam)]

for station_name, temp, humidity, rainfall in stations:
    temp_float = np.array(temp, dtype=float)
    humidity_float = np.array(humidity, dtype=float)
    rainfall_float = np.array(rainfall, dtype=float)

    dates = dates_bac if station_name == "BẮC" else dates_nam
    
    print(f"\nTRẠM {station_name}:")
 
    print("  Nhiệt độ:")
    print(f"    Trung bình: {np.mean(temp_float):.2f}°C")
    print(f"    Độ lệch chuẩn: {np.std(temp_float):.2f}")
    print(f"    Cao nhất: {np.max(temp_float):.2f}°C - Ngày: {dates[np.argmax(temp_float)]}")
    print(f"    Thấp nhất: {np.min(temp_float):.2f}°C - Ngày: {dates[np.argmin(temp_float)]}")

    print("  Độ ẩm:")
    print(f"    Trung bình: {np.mean(humidity_float):.2f}%")
    print(f"    Độ lệch chuẩn: {np.std(humidity_float):.2f}")
    print(f"    Cao nhất: {np.max(humidity_float):.2f}% - Ngày: {dates[np.argmax(humidity_float)]}")
    print(f"    Thấp nhất: {np.min(humidity_float):.2f}% - Ngày: {dates[np.argmin(humidity_float)]}")
   
    print("  Lượng mưa:")
    print(f"    Trung bình: {np.mean(rainfall_float):.2f}mm")
    print(f"    Độ lệch chuẩn: {np.std(rainfall_float):.2f}")
    print(f"    Cao nhất: {np.max(rainfall_float):.2f}mm - Ngày: {dates[np.argmax(rainfall_float)]}")
    print(f"    Thấp nhất: {np.min(rainfall_float):.2f}mm - Ngày: {dates[np.argmin(rainfall_float)]}")
```

Nhiệm vụ 11: So sánh tổng quan 6 tháng giữa hai trạm.

```python
temp_bac_f, humidity_bac_f, rainfall_bac_f = np.array(temp_bac, dtype=float), np.array(humidity_bac, dtype=float), np.array(rainfall_bac, dtype=float)
temp_nam_f, humidity_nam_f, rainfall_nam_f = np.array(temp_nam, dtype=float), np.array(humidity_nam, dtype=float), np.array(rainfall_nam, dtype=float)

stats = [
    ("Nhiệt độ TB (°C)", np.mean(temp_bac_f), np.mean(temp_nam_f), "°C"),
    ("Tổng lượng mưa (mm)", np.sum(rainfall_bac_f), np.sum(rainfall_nam_f), "mm"),
    ("Độ ẩm TB (%)", np.mean(humidity_bac_f), np.mean(humidity_nam_f), "%"),
    ("Ngày nóng (>35°C)", np.sum(temp_bac_f > 35), np.sum(temp_nam_f > 35), "ngày"),
    ("Ngày mưa lớn (>50mm)", np.sum(rainfall_bac_f > 50), np.sum(rainfall_nam_f > 50), "ngày")
]

print("\n BẢNG SO SÁNH:")
print("=" * 60)
print(f"{'CHỈ SỐ':<25} {'TRẠM BẮC':<12} {'TRẠM NAM':<12} {'SO SÁNH':<10}")
print("=" * 60)

for name, bac, nam, i in stats:
    compare = "Nam > Bắc" if nam > bac else "Bắc > Nam"
    if name == "Độ dao động nhiệt":
        compare = "Bắc ổn định hơn" if bac < nam else "Nam ổn định hơn"
    
    if i:
        print(f"{name:<25} {bac:<12.2f} {nam:<12.2f} {compare:<10}")
    else:
        print(f"{name:<25} {bac:<12.2f} {nam:<12.2f} {compare:<10}")

print("=" * 60)

print("\nKẾT LUẬN:")
print("-" * 40)
Ket_Luan = [
    f"Trạm {'Nam' if np.mean(temp_nam_f) > np.mean(temp_bac_f) else 'Bắc'} có nhiệt độ trung bình cao hơn",
    f"Trạm {'Nam' if np.sum(rainfall_nam_f) > np.sum(rainfall_bac_f) else 'Bắc'} có tổng lượng mưa cao hơn",
    f"Trạm {'Nam' if np.sum(temp_nam_f > 35) > np.sum(temp_bac_f > 35) else 'Bắc'} có nhiều ngày nóng hơn"
]

for KetLuan in Ket_Luan:
    print(KetLuan)
```

Nhiệm vụ 12: Tạo mảng tổng hợp và lưu file ket_qua_case_6.csv

```python
def phan_loai_ngay(temp, humidity, rainfall):
    n = len(temp)
    loai = np.empty(n, dtype='U20')

    temp_diff = np.abs(np.diff(temp))
    humidity_diff = np.abs(np.diff(humidity))
    rainfall_diff = np.abs(np.diff(rainfall))
    
    for i in range(n):
        if i > 0:
            if temp_diff[i-1] > 8 or humidity_diff[i-1] > 15 or rainfall_diff[i-1] > 40:
                loai[i] = "Bien dong manh"
                continue

        if temp[i] > 35 or humidity[i] < 30 or rainfall[i] > 50:
            loai[i] = "Canh bao"
        else:
            loai[i] = "Binh thuong"
    
    return loai

temp_bac_f = np.array(temp_bac, dtype=float)
humidity_bac_f = np.array(humidity_bac, dtype=float)
rainfall_bac_f = np.array(rainfall_bac, dtype=float)

temp_nam_f = np.array(temp_nam, dtype=float)
humidity_nam_f = np.array(humidity_nam, dtype=float)
rainfall_nam_f = np.array(rainfall_nam, dtype=float)

loai_bac = phan_loai_ngay(temp_bac_f, humidity_bac_f, rainfall_bac_f)
loai_nam = phan_loai_ngay(temp_nam_f, humidity_nam_f, rainfall_nam_f)

all_data = []
all_data.append(["THONG KE TONG HOP"])
all_data.append(["Tram", "Nhiet_do_TB", "Nhiet_do_Std", "Nhiet_do_Max", "Nhiet_do_Min",
                "Do_am_TB", "Do_am_Std", "Do_am_Max", "Do_am_Min", 
                "Mua_TB", "Mua_Std", "Mua_Max", "Mua_Min", "Tong_mua",
                "So_ngay_nong", "So_ngay_mua_lon"])

for station_name, temperature, humidity, rainfall in [("BAC", temp_bac, humidity_bac, rainfall_bac),
                                                      ("NAM", temp_nam, humidity_nam, rainfall_nam)]:
    
    temp_f = np.array(temperature, dtype=float)
    humidity_f = np.array(humidity, dtype=float)
    rainfall_f = np.array(rainfall, dtype=float)
    
    station_stats = [
        station_name,
        f"{np.mean(temp_f):.2f}", f"{np.std(temp_f):.2f}", f"{np.max(temp_f):.2f}", f"{np.min(temp_f):.2f}",  
        f"{np.mean(humidity_f):.2f}", f"{np.std(humidity_f):.2f}", f"{np.max(humidity_f):.2f}", f"{np.min(humidity_f):.2f}",  
        f"{np.mean(rainfall_f):.2f}", f"{np.std(rainfall_f):.2f}", f"{np.max(rainfall_f):.2f}", f"{np.min(rainfall_f):.2f}", f"{np.sum(rainfall_f):.2f}", 
        str(np.sum(temp_f > 35)),  
        str(np.sum(rainfall_f > 50))  
    ]
    
    all_data.append(station_stats)

all_data.append([])
all_data.append(["CHI TIET TUNG NGAY"])
all_data.append(["date", "temp_bac", "temp_nam", "humidity_bac", "humidity_nam", 
                 "rainfall_bac", "rainfall_nam", "loai_bac", "loai_nam"])

n = len(dates_bac)
for i in range(n):
    row = [
        dates_bac[i],
        f"{temp_bac_f[i]:.2f}", f"{temp_nam_f[i]:.2f}",
        f"{humidity_bac_f[i]:.2f}", f"{humidity_nam_f[i]:.2f}",
        f"{rainfall_bac_f[i]:.2f}", f"{rainfall_nam_f[i]:.2f}",
        loai_bac[i], loai_nam[i]
    ]
    all_data.append(row)

with open('ket_qua_case_6.csv', 'w', encoding='utf-8') as f:
    for row in all_data:
        f.write(','.join(row) + '\n')

print("Đã lưu file: 'ket_qua_case_6.csv'")
print("  File bao gồm:")
print("  - Bang thong ke tong hop (2 dong)")
print("  - Bang chi tiet tung ngay (180 dong)")

print("\nTHONG KE PHAN LOAI:")
for name, loai in [("Tram Bac", loai_bac), ("Tram Nam", loai_nam)]:
    loai_arr = np.array(loai)
    print(f"\n{name}:")
    print(f"  Binh thuong: {np.sum(loai_arr == 'Binh thuong')} ngay")
    print(f"  Canh bao: {np.sum(loai_arr == 'Canh bao')} ngay")
    print(f"  Bien dong manh: {np.sum(loai_arr == 'Bien dong manh')} ngay")
```
